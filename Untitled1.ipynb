```python
import numpy as np

K = 1.5
columns = ["fixed acidity", "volatile acidity", "chlorides", "density","sulphates","total sulfur dioxide","citric acid"]

for column in columns:
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  top_margin = Q3 + IQR * K
  low_margin = Q1 - IQR * K

  data[column] = data[column].apply(lambda x: data[column].median() if x < low_margin or x > top_margin else x)
```

# Realizamos la normalización
Escogemos los campos que no sigan una distribucion normal

```python
data.hist(bins=50,figsize=(10,10))
```

Los campos que cumplen con una distribucion normal son sulphates,density y chlorides, por lo que los demás pasarán a ser tratados con un min-max scaler

```python
from sklearn.preprocessing import MinMaxScaler
mm_scaler =  MinMaxScaler()
mm_scaler.fit(data.drop(columns=["quality","density","chlorides"]))
data_norm = mm_scaler.transform(data.drop(columns=["quality","density","chlorides"]))

prueba = pd.DataFrame(data=data_norm,columns=data.drop(columns=["quality","density","chlorides"]).columns)
data[prueba.columns] = data_norm
data.hist(bins=50,figsize=(10,10))
```

Con esto conseguimos mantener una "relacion" con las caracteristicas que siguen una distribucion normal evitando errores a la hora de entrenar, IMPORTANTE esto es normalizacion, ya que si miramos un histogramas los datos que no siguen una normal tienen valores similares a los que si siguen una normal, es por ello que hay que escalarlos TODOS con Standard Scaler

# DESBALANCEO DE DATOS

```python
from imblearn.over_sampling import SMOTE,RandomOverSampler # SMOTE - genera datos sinteticos para balancear los datos  
                                                           # RandomOverSampler - duplica los ejemplos existentes  
rus = SMOTE()
quality = data["quality"]

caract_X, caract_Y = rus.fit_resample(data.drop(columns=["quality"]),quality)

print(f"Datos originales {len(quality)} - datos balanceados {len(caract_Y)}")
caract_X["quality"] = caract_Y
caract_X.hist(bins=50,figsize=(10,10))
```
